```python
#From https://stackoverflow.com/a/21894086
class bidict(dict):
    def __init__(self, *args, **kwargs):
        super(bidict, self).__init__(*args, **kwargs)
        self.inverse = {}
        for key, value in self.items():
            self.inverse.setdefault(value,[]).append(key) 

    def __setitem__(self, key, value):
        if key in self:
            self.inverse[self[key]].remove(key) 
        super(bidict, self).__setitem__(key, value)
        self.inverse.setdefault(value,[]).append(key)        

    def __delitem__(self, key):
        self.inverse.setdefault(self[key],[]).remove(key)
        if self[key] in self.inverse and not self.inverse[self[key]]: 
            del self.inverse[self[key]]
        super(bidict, self).__delitem__(key)
```

```python
import copy

#make a function called "shift_key" 
#takes two arguments: degree_map, new_key
#return new degree map.

def shift_key(input_degree_map,new_key):
    new_degree_map = copy.copy(input_degree_map)
    for k,v in input_degree_map.items():
        new_degree_map[(k-input_degree_map.inverse[new_key][0])%12] = v
    return new_degree_map
```

```python
degree_map = bidict({0:'C',1:'C#',2:'D',3:'D#',4:'E',5:'F',6:'F#',7:'G',8:'G#',9:'A',10:'A#',11:'B'})
a_map = shift_key(degree_map,'A')
```

```python
#Function that takes a midi int note, and prints jianpu dots:
def convert_notes(midi_int_note):
    dot_discriminator = (midi_int_note//12)-3
    if dot_discriminator >=0:
        n_dots = dot_discriminator
        for _ in range(n_dots):
            print('.')
    print(a_map[midi_int_note%12])
    if dot_discriminator<0:
        n_dots = abs(dot_discriminator)
        for _ in range(n_dots):
            print('.')
```
